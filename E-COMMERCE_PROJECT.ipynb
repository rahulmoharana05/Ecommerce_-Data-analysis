# converting the date data type as object to date 

```python
dataset["Order Date"]=pd.to_datetime(dataset["Order Date"])
```

```python
dataset.info()
```

```python
dataset["Ship Date"]=pd.to_datetime(dataset["Ship Date"])
```

```python
dataset["order month"]=dataset['Order Date'].dt.month
dataset['Order year']=dataset['Order Date'].dt.year
dataset['Order day of week']=dataset['Order Date'].dt.dayofweek
dataset['days to ship']=(dataset['Ship Date']-dataset['Order Date']).dt.days
```

# profit and sales by customer segement

```python
sales_profit_by_segement=dataset.groupby('Segment')[['Sales','Profit']].sum().reset_index()
```

```python
sales_profit_by_segement
```

```python

melted_df = sales_profit_by_segement.melt(id_vars='Segment', value_vars=['Sales', 'Profit'],
                                         var_name='Metric', value_name='Amount')

# Create the barplot
sns.barplot(x='Segment', y='Amount', hue='Metric', data=melted_df)

# Add titles and labels
plt.title('Sales and Profit by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Amount')
plt.legend(title='Metric')

# Show the plot
plt.show()
```

# sales to profit ratio

```python
sales_profit_ratio_by_segment=dataset.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sales_profit_ratio_by_segment['sales_to_profit_ratio']=sales_profit_ratio_by_segment['Sales']/sales_profit_ratio_by_segment['Profit']
print(sales_profit_ratio_by_segment[["Segment",'sales_to_profit_ratio']])
```

```python
sales_profit_ratio_by_segment
```

```python
plt.pie(sales_profit_ratio_by_segment['sales_to_profit_ratio'],labels=sales_profit_by_segement['Segment'])
plt.show()
```
